```python
## Splitting the dataset into dependent and independent variables
```

```python
x = dataset_df.iloc[:, :-1].values
y = dataset_df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
import pickle
pickle.dump(sc, open('scaler.pkl', 'wb'))
```

```python
## Training the model ##
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_model.fit(x_train, y_train)
```

```python
y_pred_knn = knn_model.predict(x_test)
```

```python
report_knn = classification_report(y_test, y_pred_knn)
print(report_knn)
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='linear')
svc_model.fit(x_train, y_train)
```

```python
y_pred_svc = svc_model.predict(x_test)
```

```python
report_svc = classification_report(y_test, y_pred_svc)
print(report_svc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
report_rf = classification_report(y_test, y_pred_rf)
print(report_rf)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion='entropy')
dt_model.fit(x_train, y_train)
```

```python
y_pred_dt = dt_model.predict(x_test)
```

```python
report_dt = classification_report(y_test, y_pred_dt)
print(report_dt)
```

```python
## Hyper turning the model ##
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [10,20,30,50, 100,150, 200],    # Number of trees
    'max_depth': [None,5, 10, 15, 20, 25],       # Depth of trees
    'min_samples_split': [2, 5, 10, 12, 15],   # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 3, 4, 5]      # Minimum samples at leaf node
}

gridCV = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, verbose=1, cv=5, scoring='accuracy', n_jobs=-1)
gridCV.fit(x_train, y_train)
```

```python
print("Best Hyperparameters:", gridCV.best_params_)
print("Best Accuracy:", gridCV.best_score_)
print("Best Estimator:", gridCV.best_estimator_)
print("Best Index:", gridCV.best_index_)
print("Scorer:", gridCV.scorer_)
print("CV Results:", gridCV.cv_results_)
```

```python
rf_model_best = RandomForestClassifier(max_depth=20, min_samples_leaf=3, min_samples_split=2, n_estimators=30)
rf_model_best.fit(x_train, y_train)
```

```python
rf_model_best_pred = rf_model_best.predict(x_test)
report_rf_best = classification_report(y_test, rf_model_best_pred)
print(report_rf_best)
```
