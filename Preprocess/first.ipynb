```python
# Importing the required libraries
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Importing the dataset
```

```python
dataset = pd.read_csv('data.csv')
dataset.head()
```

```python
#Creating the dependent and independent variables
```

```python
dataset_df = dataset.copy()
```

```python
# HANDLING MISSING DATA
```

```python
missing_values = dataset_df.isnull().sum()
print(missing_values)
```

```python
import seaborn as sns

# Visualizing the missing values
plt.figure(figsize=(10, 6))
sns.heatmap(dataset_df.isnull(), cbar=True, cmap='coolwarm')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Visual key distribution of the dataset
```

```python
dataset_df.hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.suptitle('Distribution of All Features')
plt.show()
```

```python
# Set figure size
area = plt.figure(figsize=(16, 16))

#  Age 
plt.subplot(4, 3, 1)
sns.histplot(dataset_df['age'], bins=20, kde=True, color="skyblue", edgecolor="black" , label = 'Age')
plt.xlabel("Age")
plt.title("Age Distribution")
plt.legend(loc="upper right", fontsize='small')

# Sex Distribution  
plt.subplot(4, 3, 2)
dataset_df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=["lightblue", "salmon"] , label = 'sex')
plt.ylabel("")
plt.title("Sex Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Chest Pain Type (cp)   
plt.subplot(4, 3, 3)
sns.violinplot(x=dataset_df['cp'], color="lightgreen" , label = 'Chest Pain Type')
plt.xlabel("Chest Pain Type (cp)")
plt.title("Chest Pain Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Blood Pressure (trestbps)  
plt.subplot(4, 3, 4)
sns.boxplot(x=dataset_df['trestbps'], color="red" , label = 'Resting Blood Pressure')
plt.xlabel("Resting Blood Pressure")
plt.title("Blood Pressure Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Cholesterol (chol)  
plt.subplot(4, 3, 5)
sns.kdeplot(dataset_df['chol'], fill=True, color="purple" ,  label = 'Cholesterol Level')
plt.xlabel("Cholesterol Level")
plt.title("Cholesterol Distribution")
plt.legend(loc="upper right", fontsize='small')

# Fasting Blood Sugar (fbs)  
plt.subplot(4, 3, 6)
dataset_df['fbs'].value_counts().plot(kind='bar', color="lightblue" ,   label = 'Fasting Blood Sugar')
plt.xlabel("Fasting Blood Sugar")
plt.title("FBS Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Resting ECG (restecg)   
plt.subplot(4, 3, 7)
dataset_df['restecg'].value_counts().plot(kind='bar', color="coral" ,   label = 'Rest ECG')
plt.xlabel("Rest ECG")
plt.title("Rest ECG Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Maximum Heart Rate (thalach)  
plt.subplot(4, 3, 8)
sns.histplot(dataset_df['thalach'], bins=20, kde=True, color="red" , edgecolor="black" , label = 'Max Heart Rate')
plt.xlabel("Max Heart Rate")
plt.title("Max Heart Rate Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Exercise-Induced Angina (exang)  
plt.subplot(4, 3, 9)
dataset_df['exang'].value_counts().plot(kind='bar', color="orange" ,  label = 'Exercise-Induced Angina')
plt.xlabel("Exercise-Induced Angina")
plt.title("Exang Distribution")
plt.legend(loc="upper right", fontsize='small')

# Oldpeak 
plt.subplot(4, 3, 10)
sns.stripplot(x=dataset_df['oldpeak'], color="cyan", jitter=True , label = 'Oldpeak')
plt.xlabel("Oldpeak")
plt.title("Oldpeak Distribution")
plt.legend(loc="upper right", fontsize='small')

# Slope  
plt.subplot(4, 3, 11)
dataset_df['slope'].value_counts().plot(kind='bar', color="brown" , label = 'Slope')
plt.xlabel("Slope")
plt.title("Slope Distribution")
plt.legend(loc="upper right", fontsize='small')

#  Major Vessels (ca)  
plt.subplot(4, 3, 12)
sns.swarmplot(x=dataset_df['ca'], color="gold"  , label = 'Major Vessels')
plt.xlabel("Number of Major Vessels (ca)")
plt.title("Major Vessels Distribution")
plt.legend(loc="upper right", fontsize='small')

plt.tight_layout()
plt.show()
```

```python
# Heart Disease Frequency According to age
plt.figure(figsize=(8, 5))
sns.histplot(dataset_df, x="age", hue="condition", kde=True, bins=20, palette=["yellow", "darkblue"])
plt.title("Age Distribution Based on Heart Disease", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend(title="Heart Disease", labels=["No", "Yes"])
plt.show()
```

```python
# Get min, max, and average age of people who do not have heart disease
no_disease_ages = dataset_df[dataset_df['condition'] == 0]['age']
print(f"Min age of people without heart disease: {no_disease_ages.min()}")
print(f"Max age of people without heart disease: {no_disease_ages.max()}")
print(f"Average age of people without heart disease: {no_disease_ages.mean():.2f}")
```

```python
# ST depression distribution based on heart disease
fig, axes = plt.subplots(1, 2, figsize=(20, 5))
sns.histplot(dataset_df[dataset_df['condition'] == 0]['oldpeak'], bins=20, kde=True, color="blue", ax=axes[0])
axes[0].set_title("ST Depression Distribution (No Heart Disease)", fontsize=14)
axes[0].set_xlabel("ST Depression")
axes[0].set_ylabel("Frequency")
sns.histplot(dataset_df[dataset_df['condition'] == 1]['oldpeak'], bins=20, kde=True, color="green", ax=axes[1])
axes[1].set_title("ST Depression Distribution (Heart Disease)", fontsize=14)
axes[1].set_xlabel("ST Depression")
axes[1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Heart Disease Frequency According to slope of the peak exercise ST segment
dataset_df['slope'] = dataset_df['slope'].map({0: 'upsloping', 1: 'flat', 2: 'downsloping'})
plt.figure(figsize=(8, 5))
sns.countplot(data=dataset_df, x="slope", hue="condition", palette=["magenta", "lime"])
plt.title('Heart Disease Frequency According to Slope of the Peak Exercise ST Segment')
plt.xlabel('Slope of the Peak Exercise ST Segment')
plt.ylabel('Frequency')
plt.legend(title="Heart Disease", labels=["No", "Yes"])
plt.show()
```

```python
# Heart Disease Frequency According to Number of vessels based on the target
plt.figure(figsize=(8, 5))
sns.countplot(data=dataset_df, x="ca", hue="condition", palette=["teal", "coral"])
plt.title('Heart Disease Frequency According to Number of Major Vessels')
plt.xlabel('Number of Major Vessels')
plt.ylabel('Frequency')
plt.legend(title="Heart Disease", labels=["No", "Yes"])
plt.show()
```

```python
# Heart Disease Frequency According to Thalassemia
dataset_df['thal'] = dataset_df['thal'].map({0: 'normal', 1: 'fixed defect', 2: 'reversable defect'})
plt.figure(figsize=(8, 5))
sns.countplot(data=dataset_df, x="thal", hue="condition", palette=["purple", "orange"])
plt.title('Heart Disease Frequency According to Thalassemia')
plt.xlabel('Thalassemia')
plt.ylabel('Frequency')
plt.legend(title="Heart Disease", labels=["No", "Yes"])
plt.show()
```

```python
# Correlation matrix 
corr_matrix = dataset_df.select_dtypes(include=['int64', 'float64']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
